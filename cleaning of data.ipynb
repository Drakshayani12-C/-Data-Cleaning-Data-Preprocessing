```python
# Fill missing values with median for selected columns
for i in ["Age", "Fare"]:
    df[i].fillna(df[i].median(), inplace=True)

# Import and initialize the KNNImputer
from sklearn.impute import KNNImputer
impute = KNNImputer()

# Apply KNNImputer to numerical columns
for i in df.select_dtypes(include="number").columns:
    df[i] = impute.fit_transform(df[[i]])

# Check remaining missing values
df.isnull().sum()
```

```python
def wisker(col):
    q1,q3=np.percentile(col,[25,75])
    iqr=q3-q1
    lw=q1-1*5*iqr
    uw=q3+1*5*iqr
    return lw,uw
```

```python
# List of numeric columns you want to clean
cols = ['Age', 'Fare']  # Add more if needed
```

```python
# Apply wisker and replace outliers
for i in cols:
    lw, uw = wisker(df[i])
    df[i] = np.where(df[i] < lw, lw, df[i])
    df[i] = np.where(df[i] > uw, uw, df[i])
```

```python
# Plot boxplots for cleaned columns
for i in cols:
    sns.boxplot(df[i])
    plt.title(f'Boxplot of {i}')
    plt.show()
```
